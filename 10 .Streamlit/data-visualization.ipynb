```python
from wordcloud import WordCloud
```

```python
wordcloud_df['features'] = wordcloud_df['features'].str.strip("""['"]""").str.split("', '")
```

```python
def sector_wise(df):
    sectors = df['sector'].unique()
    new_df = []
    def foo(x):
        words.extend(x)
    for sector in sectors:
        words = []
        df[df['sector']==sector]['features'].dropna().apply(foo)
        new_df.append([sector,words])
    return pd.DataFrame(new_df,columns=['sector','words'])
```

```python
words = sector_wise(wordcloud_df)
```

```python
import pickle
```

```python
with open('words.pkl','wb') as f:
    pickle.dump(words,f)
```

```python
text = words[words['sector']=='sector 36']['words'][0]
feature_text = ' '.join(text)
```

```python
feature_text
```

```python
plt.rcParams["font.family"] = "Arial"

wordcloud = WordCloud(width = 800, height = 800, 
                      background_color ='white', 
                      stopwords = set(['s']),  # Any stopwords you'd like to exclude
                      min_font_size = 10).generate(feature_text)

plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud, interpolation='bilinear') 
plt.axis("off") 
plt.tight_layout(pad = 0) 
plt.show() # st.pyplot()
```
