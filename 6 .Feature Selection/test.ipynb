```python
import numpy as np
import pandas as pd
import os
```

```python
df = pd.read_csv('../5 .Missing Value Treatment/gurgaon_properties_missing_value_imputation.csv')
```

```python
df.head()
```

```python
X = df.drop(columns=['society','price_per_sqft','price'])
Y = df['price']
```

```python
X.shape
```

```python
def categorize_luxury(score):
    if 0 <= score < 50:
        return "Low"
    elif 50 <= score < 150:
        return "Medium"
    elif 150 <= score <= 175:
        return "High"
    else:
        return None  # or "Undefined" or any other label for scores outside the defined bins
    
X['luxury_category'] =X['luxury_score'].apply(categorize_luxury)
```

```python
def categorize_floor(floor):
    if 0 <= floor <= 2:
        return "Low Floor"
    elif 3 <= floor <= 10:
        return "Mid Floor"
    elif 11 <= floor <= 51:
        return "High Floor"
    else:
        return None  # or "Undefined" or any other label for floors outside the defined bins
X['floor_category'] = X['floorNum'].apply(categorize_floor)
```

```python
X.drop(columns=['floorNum','luxury_score'],inplace=True)
```

```python
from sklearn.preprocessing import OrdinalEncoder

# Create a copy of the original data for label encoding

categorical_cols = X.select_dtypes(include=['object']).columns

# Apply label encoding to categorical columns
for col in categorical_cols:
    oe = OrdinalEncoder()
    X[col] = oe.fit_transform(X[[col]])
    print(oe.categories_)
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test =  train_test_split(X,Y)
```

# T3: Mutual Information

# T4:Tree Based Model and Lasso

```python
rf = RandomForestRegressor()
rf.fit(X_train,y_train)
t1 = pd.DataFrame({'columns':X_train.columns,'forest':rf.feature_importances_})
t1 = t1.sort_values(by=['forest'],ascending=False)
t1
```

# Lasso

```python
from sklearn.linear_model import Lasso
lr = Lasso(alpha=0.01)
lr.fit(X_train,y_train)
t2 = pd.DataFrame({'columns':X_train.columns,'Lasso':abs(lr.coef_)})
t2.sort_values(by=['Lasso'],ascending=False)
t2
```

# T5 : RFE

```python
from sklearn.feature_selection import RFECV
rf = RandomForestRegressor()
rfe = RFECV(rf,min_features_to_select=X_train.shape[1],cv=5)
rfe.fit(X_train,y_train)
rfe.get_support()
```

```python
t3 = pd.DataFrame({'columns':X_train.columns,'rfe':rfe.estimator_.feature_importances_})
t3 = t3.sort_values(by=['rfe'],ascending=False)
t3
```

# T6. Sequential Selection


```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
rf = RandomForestRegressor()
sfc = SFS(rf,k_features='best',forward=False,floating=False,scoring='r2')
model = sfc.fit(X_train,y_train)
```

```python
model.k_feature_idx_
```

```python
model.k_score_
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
rf = RandomForestRegressor()
rf.fit(X_train.iloc[:,list(model.k_feature_idx_)],y_train)
y_pred = rf.predict(X_test.iloc[:,list(model.k_feature_idx_)])
r2_score(y_true=y_test,y_pred=y_pred)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
rf = LinearRegression()
rf.fit(X_train.iloc[:,list(model.k_feature_idx_)],y_train)
y_pred = rf.predict(X_test.iloc[:,list(model.k_feature_idx_)])
r2_score(y_true=y_test,y_pred=y_pred)
```
